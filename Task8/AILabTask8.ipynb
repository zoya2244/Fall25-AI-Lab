```python
import math

values = []        
depthLimit = 0      

def miniMaxAlgo():
    def traverse(level, node, isMaxTurn):
        if level == depthLimit:
            return values[node]

        if isMaxTurn:
            print(f"Maximizer at level {level}, node {node}")
            leftVal = traverse(level + 1, node * 2, False)
            rightVal = traverse(level + 1, node * 2 + 1, False)
            return max(leftVal, rightVal)
        else:
            print(f"Minimizer at level {level}, node {node}")
            leftVal = traverse(level + 1, node * 2, True)
            rightVal = traverse(level + 1, node * 2 + 1, True)
            return min(leftVal, rightVal)

    return traverse(0, 0, True)


print("==== MINIMAX ALGORITHM ====\n")

values = list(map(int, input("Enter leaf node values: ").split()))
depthLimit = int(math.log2(len(values)))

print("\nProcessing...\n")

bestScore = miniMaxAlgo()

print("\n------------------------------------")
print(f"The Optimal Result for Maximizer: {bestScore}")
print("------------------------------------")
```
