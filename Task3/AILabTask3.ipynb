```python
 Model-Based Reflex Agent 
# This agent not only checks the current temperature but also remembers the previous action to avoid turning the heater on or off unnecessarily.
```

```python
class SimpleReflexAgent:
    def __init__(self, temp):
        self.fixed_temp = temp
        self.previous_action = {}  

    def sensor(self, room, temp):
        self.room = room
        self.current_temp = temp
        if room not in self.previous_action:
            self.previous_action[room] = "AC OFF"  

    def performance(self):
        action = "-"  
        prev = self.previous_action[self.room]

        if self.current_temp < self.fixed_temp and prev != "AC ON":
            action = "Turn ON AC"
            self.previous_action[self.room] = "AC ON"
        elif self.current_temp >= self.fixed_temp and prev != "AC OFF":
            action = "Turn OFF AC"
            self.previous_action[self.room] = "AC OFF"
        return action

    def actuator(self):
        action = self.performance()
        print(self.room, self.current_temp, "->", action)

# Room temperatures
rooms = {
    "Living Room": 20,
    "Drawing Room": 22,
    "Kitchen": 24,
}

agent = SimpleReflexAgent(22)  


print( "Check")
for room, temp in rooms.items():
    agent.sensor(room, temp)
    agent.actuator()


updated_rooms = {
    "Living Room": 23,
    "Drawing Room": 21,
    "Kitchen": 22,
}

print("\nCheck Again")
for room, temp in updated_rooms.items():
    agent.sensor(room, temp)
    agent.actuator()
```
