# BFS without queue

```python
def bfs_without_queue(graph, start):
    visited = []

    def bfs_level(level):
        if not level:
            return
        nxt_level = []
        for bfs in level:
            if bfs not in visited:
                print(bfs, end=" ")
                visited.append(bfs)
                nxt_level.extend(graph[bfs])
        bfs_level(nxt_level)

    bfs_level([start])
graph = {
   'A': ['B', 'C'],
   'B': ['D', 'E'],
   'C': ['F'],
   'D': ['G'],
   'E': [],
   'F': ['H'],
   'G': [],
   'H': [],
}
print("BFS without Queue:")
bfs_without_queue(graph, 'A')
```

```python
def bfs_with_queue(graph, start):
    visited = []
    i = [start]   
    while i:
        bfs = i.pop(0)   # remove first element
        if bfs not in visited:
            print(bfs, end=" ")
            visited.append(bfs)
            for j in graph[bfs]:
                if j not in visited:
                    i.append(j)

print("\nBFS with Queue:")
bfs_with_queue(graph, 'A')
```
