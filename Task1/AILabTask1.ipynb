```python
tasks = []
def addTask():
    task = input("Enter your task: ")
    tasks.append(task)
    print(f"Your Task '{task}'is added to list")
    

def listTasks():
    if len(tasks) == 0:
        print("No Tasks")
    else:
        print("No of Tasks: ")
        for index, task in enumerate (tasks, start= 1):
            print(f"Task {index}. {task}")

    
def deleteTask():
    listTasks()
    try:
        delete_task = int(input("Enter the task number to delete: "))
        if delete_task >=0 and delete_task < len(tasks):
            tasks.pop(delete_task)
            print(f"Task number {delete_task} Deleted")
        else:    
            print(f"Task number {delete_task} dont exist")
        
        
    except:
        print("Invalid.")
        
        
        
if __name__ == '__main__':
     
    print("Welcome to the To Do List ")
    while True:
        print("\n")
        print("Please select an options: ")
        print("1. Add new task")
        print("2. Delete task")
        print("3. List tasks")
        print("4. Quit")
        
        choice= input("Enter your choice: ")
        
        if choice == "1":
            addTask()
        
        elif choice == "2":
            deleteTask()
    
        elif choice == "3":
            listTasks()
            
        elif choice == "4":
            break
        else:
            print("Invalid input,Try again")
```
