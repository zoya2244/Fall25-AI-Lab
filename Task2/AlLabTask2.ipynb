```python
print(" FizzBuzz\n")

previous_value = 0  

for comp_number in range(1, 100):  
    print(f"Computer : {comp_number}")

    if previous_value == 0:
        print("No previous value.\n")
        previous_value = comp_number
        continue

    rule = previous_value + comp_number
    print(f"Rule = {previous_value} + {comp_number} = {rule}")

    if rule % 3 == 0:
        print("Fizz\n")
    elif rule % 5 == 0:
        print("Buzz\n")
    elif rule % 3 == 0 and rule % 5 == 0:
        print("FizzBuzz\n")
    else:
        print(f"{rule}\n")

    previous_value = comp_number
```
