```python
def luhn_check(card_number):
    card_number = card_number[::-1] 
    total = 0

    for i, digit in enumerate(card_number):
        num = int(digit)  
        if i % 2 == 1:
            num = num * 2
            
            if num > 9:
                num -= 9
        total += num
    return total % 10 == 0
card = input("Enter a card number to validate: ")

if card.isdigit():
    if luhn_check(card):
        print("Valid card number (Luhn check passed).")
    else:
        print("Invalid card number (Luhn check failed).")
else:
    print("Please enter digits only.")
```
